```python
import pandas as pd
```

```python
rand_pred_eng = pd.read_csv('/content/probing_compositionality/probing/Russian_English/results/res_rand_pred_eng.csv')
test_cls_eng = pd.read_csv('/content/probing_compositionality/probing/Russian_English/results/res_test_cls_eng.csv')
train_cls_eng = pd.read_csv('/content/probing_compositionality/probing/Russian_English/results/res_train_cls_eng.csv')
test_rand_vec_eng = pd.read_csv('/content/probing_compositionality/probing/Russian_English/results/res_test_rand_vec_eng.csv')
train_rand_vec_eng = pd.read_csv('/content/probing_compositionality/probing/Russian_English/results/res_train_rand_vec_eng.csv')
train_mean_eng = pd.read_csv('/content/probing_compositionality/probing/Russian_English/results/res_train_mean_eng.csv')
test_mean_eng = pd.read_csv('/content/probing_compositionality/probing/Russian_English/results/res_test_mean_eng.csv')
```

```python
all_test_eng = [rand_pred_eng.iloc[:, 2:].mean(), test_rand_vec_eng.iloc[:, 2:].mean(),
            test_cls_eng.iloc[:, 2:].mean(), test_mean_eng.iloc[:, 2:].mean()]
all_train_eng = [train_rand_vec_eng.iloc[:, 2:].mean(), train_cls_eng.iloc[:, 2:].mean(), train_mean_eng.iloc[:, 2:].mean()]
```

```python
pd.concat(all_test_eng, axis=1).rename(columns={0:'rand_pred_eng', 1:'test_rand_vec_eng', 2:'test_cls_eng', 3:'test_mean_eng'})
```

```python
pd.concat(all_train_eng, axis=1).rename(columns={0:'train_rand_vec_eng', 1:'train_cls_eng', 2:'train_mean_eng'})
```
