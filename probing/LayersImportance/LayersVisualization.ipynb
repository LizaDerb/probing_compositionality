```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
eng_bert_cls = pd.read_csv('/content/probing_compositionality/probing/LayersImportance/English_BERT/data_files/res_test_cls.csv')['f1']
eng_bert_mean = pd.read_csv('/content/probing_compositionality/probing/LayersImportance/English_BERT/data_files/res_test_mean.csv')['f1']
rus_bert_cls = pd.read_csv('/content/probing_compositionality/probing/LayersImportance/Russian_RuBERT/data_files/res_test_cls.csv')['f1']
rus_bert_mean = pd.read_csv('/content/probing_compositionality/probing/LayersImportance/Russian_RuBERT/data_files/res_test_mean.csv')['f1']
eng_mbert_cls = pd.read_csv('/content/probing_compositionality/probing/LayersImportance/RussianEnglish_mBERT/data_files_eng/res_test_cls.csv')['f1']
eng_mbert_mean = pd.read_csv('/content/probing_compositionality/probing/LayersImportance/RussianEnglish_mBERT/data_files_eng/res_test_mean.csv')['f1']
rus_mbert_cls = pd.read_csv('/content/probing_compositionality/probing/LayersImportance/RussianEnglish_mBERT/data_files_rus/res_test_cls.csv')['f1']
rus_mbert_mean = pd.read_csv('/content/probing_compositionality/probing/LayersImportance/RussianEnglish_mBERT/data_files_rus/res_test_mean.csv')['f1']
```

```python
x = range(1, 13)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
ax[0].plot(x, eng_bert_cls, label='bert')
ax[0].plot(x, eng_mbert_cls, label='eng_mbert', linestyle="--")
ax[0].plot(x, rus_bert_cls, label='rubert', linestyle="-.")
ax[0].plot(x, rus_mbert_cls, label='rus_mbert', linestyle=":")
ax[0].set_xlabel('F1-score CLS')
ax[0].set_xticks(x, range(1, 13), rotation='horizontal')
ax[0].legend()
ax[1].plot(x, eng_bert_mean, label='bert')
ax[1].plot(x, eng_mbert_mean, label='eng_mbert', linestyle="--")
ax[1].plot(x, rus_bert_mean, label='rubert', linestyle="-.")
ax[1].plot(x, rus_mbert_mean, label='rus_mbert', linestyle=":")
ax[1].set_xlabel('F1-score Mean')
ax[1].set_xticks(x, range(1, 13), rotation='horizontal')
ax[1].legend()
fig.savefig('plots_colorful.png', bbox_inches = 'tight', dpi=500)
plt.show()
```

```python
x = range(1, 13)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
ax[0].plot(x, eng_bert_cls, label='bert', color='gray')
ax[0].plot(x, eng_mbert_cls, label='eng_mbert', linestyle="--", color='darkgray')
ax[0].plot(x, rus_bert_cls, label='rubert', linestyle="-.", color='black')
ax[0].plot(x, rus_mbert_cls, label='rus_mbert', linestyle=":", color='#778899')
ax[0].set_xlabel('F1-score CLS')
ax[0].set_xticks(x, range(1, 13), rotation='horizontal')
ax[0].legend()
ax[1].plot(x, eng_bert_mean, label='bert', color='gray')
ax[1].plot(x, eng_mbert_mean, label='eng_mbert', linestyle="--", color='darkgray')
ax[1].plot(x, rus_bert_mean, label='rubert', linestyle="-.", color='black')
ax[1].plot(x, rus_mbert_mean, label='rus_mbert', linestyle=":", color='#778899')
ax[1].set_xlabel('F1-score Mean')
ax[1].set_xticks(x, range(1, 13), rotation='horizontal')
ax[1].legend()
fig.savefig('plots_gray.png', bbox_inches = 'tight', dpi=500)
plt.show()
```
