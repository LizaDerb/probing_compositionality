```python
import pandas as pd
```

```python
rand_pred = pd.read_csv('/content/probing_compositionality/probing/russian_AN_NN/results/res_rand_pred.csv')
test_cls = pd.read_csv('/content/probing_compositionality/probing/russian_AN_NN/results/res_test_cls.csv')
train_cls = pd.read_csv('/content/probing_compositionality/probing/russian_AN_NN/results/res_train_cls.csv')
test_rand_vec = pd.read_csv('/content/probing_compositionality/probing/russian_AN_NN/results/res_test_rand_vec.csv')
train_rand_vec = pd.read_csv('/content/probing_compositionality/probing/russian_AN_NN/results/res_train_rand_vec.csv')
train_mean = pd.read_csv('/content/probing_compositionality/probing/russian_AN_NN/results/res_train_mean.csv')
test_mean = pd.read_csv('/content/probing_compositionality/probing/russian_AN_NN/results/res_test_mean.csv')
```

```python
all_test = [rand_pred.iloc[:, 2:].mean(), test_rand_vec.iloc[:, 2:].mean(),
            test_cls.iloc[:, 2:].mean(), test_mean.iloc[:, 2:].mean()]
all_train = [train_rand_vec.iloc[:, 2:].mean(), train_cls.iloc[:, 2:].mean(), train_mean.iloc[:, 2:].mean()]
```

```python
pd.concat(all_test, axis=1).rename(columns={0:'rand_pred', 1:'test_rand_vec', 2:'test_cls', 3:'test_mean'})
```

```python
pd.concat(all_train, axis=1).rename(columns={0:'train_rand_vec', 1:'train_cls', 2:'train_mean'})
```
